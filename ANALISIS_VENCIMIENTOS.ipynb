```python
#query que da el valor carga promedio por mes de segemento 
query=f'''
WITH BASE AS (
SELECT SUBSCRIBER_ID,
       MVTO,
       ANTIGUEDAD,
       SEG_RFM_MES_ACT,
       CASE
            WHEN SEG_RFM_MES_ACT LIKE '1%' THEN 'Gross'
            WHEN SEG_RFM_MES_ACT LIKE '2%' THEN 'Recurrente'
            WHEN SEG_RFM_MES_ACT LIKE '3%' THEN 'Esporadico'
            WHEN SEG_RFM_MES_ACT LIKE '4%' THEN 'Sin ingresos'
       END AS SEG_RFM_GENERAL 
FROM MVALIENTE.TBL_SEG_RFM_PREPAGO_NEW
WHERE PERIODO=202303),
BASE2 AS (
SELECT SUBSCRIBER_ID,COALESCE(ROUND(SUM(VALOR_CARGA),0),0) AS VALOR_RECARGAS
FROM DWH_BODEGA_WOM.FCT_RECARGAS
WHERE PERIODO_PROCESO_CODIGO<=202303
AND SUBSCRIBER_ID IN (
SELECT SUBSCRIBER_ID
FROM MVALIENTE.TBL_SEG_RFM_PREPAGO_NEW
WHERE PERIODO=202303
)
AND SERVICIO='Prepaid'
GROUP BY SUBSCRIBER_ID
)

SELECT A.*,
       B.VALOR_RECARGAS,
       CASE
           WHEN A.ANTIGUEDAD<1 THEN B.VALOR_RECARGAS
           ELSE ROUND(B.VALOR_RECARGAS/A.ANTIGUEDAD,0)
       END AS PROMEDIO_MES
FROM BASE A 
LEFT JOIN BASE2 B ON A.SUBSCRIBER_ID=B.SUBSCRIBER_ID
'''
promedio_carga_segmento=cst.consulta(query)
```

```python
###
query=f'''
WITH 
BASE AS(
SELECT * FROM TEMP_VENCIMIENTOS_CP ),
BASE2 AS ( 
SELECT * FROM (
SELECT SUBSCRIBER_ID,
       PREDICCION, 
       ROW_NUMBER() OVER(PARTITION BY SUBSCRIBER_ID ORDER BY FECHA_PRED DESC ) AS RN 
FROM AGG_DL_PJ_QUALITY) WHERE RN=1 
)
,
BASE3 AS (
SELECT * FROM (
SELECT B.SUBSCRIBER_ID,
       C.*,
       ROW_NUMBER() OVER(PARTITION BY SUBSCRIBER_ID ORDER BY PERIODO_PROCESO_CODIGO DESC ) AS RN  
FROM DWH_BODEGA_WOM.FCT_SUBSCRIBERS_ENDING B
LEFT JOIN DWH_BODEGA_WOM.DIM_CONSULTORES C ON B.CONSULTOR_DK=C.CONSULTOR_DK
WHERE B.MOVIMIENTO_NOMBRE='ACTIVACION'
AND B.SERVICIO='Prepaid')
WHERE RN=1
)
SELECT A.*,
        
       COALESCE(B.PREDICCION,0)AS PREDICCION,
       C.CONSULTOR_REGIONAL_NOMBRE,
       C.CONSULTOR_CANAL_NOMBRE,
       C.CONSULTOR_CIUDAD_NOMBRE,
       C.CONSULTOR_NOMBRE_NIVEL_8
        
FROM BASE A 
LEFT JOIN BASE2 B ON A.SUBS_ID=B.SUBSCRIBER_ID
LEFT JOIN BASE3 C ON A.SUBS_ID=C.SUBSCRIBER_ID
'''
vencimientos=cst.consulta(query)
```

```python
vencimientos['VENCIDO'] = vencimientos.apply(lambda row: 'NO CONSUMIDO' if -1*row['PRE_BALANCE'] == row['CHARGE'] else 'CONSUMIDO', axis=1)
```

```python

# Supongamos que tienes un DataFrame llamado "vencimientos" con las columnas "PRE_BALANCE" y "CHARGE"

def calcular_recarga_restante(row):
    pre_balance = row['PRE_BALANCE']
    charge = row['CHARGE']

    if -1*pre_balance == charge:
        return '100%'
    elif -1*0.99*pre_balance <= charge < -1*0.95*pre_balance:
        return '95%-99%'
    elif -1*0.95*pre_balance <= charge < -1*0.85*pre_balance:
        return '85%-95%'
    elif -1*0.5*pre_balance <= charge < -1*0.85*pre_balance:
        return '50%-85%'
    elif -1*0.2*pre_balance <= charge < -1*0.5*pre_balance:
        return '20%-50%'
    elif -1*0.01*pre_balance <= charge < -1*0.2*pre_balance:
        return '1%-20%'
    elif charge == 0:
        return '0%'
    else:
        return None  # Manejar otro caso si es necesario

# Crear la nueva columna "recarga_restante" usando .apply()
vencimientos['recarga_restante'] = vencimientos.apply(calcular_recarga_restante, axis=1)

```

```python
def asignar_categoria(valor):
    valor=valor*-1
    if valor >0 and valor<= 1000:
        return '0-1000'
    elif valor > 1000 and valor <=5000:
        return '1000-5000'
    elif valor >5000 and valor <=10000:
        return '5000-10000'
    else:
         return 'Mayor 10000'  
```

```python
vencimientos['CATEGORIA'] = vencimientos['PRE_BALANCE'].apply(asignar_categoria)
```
