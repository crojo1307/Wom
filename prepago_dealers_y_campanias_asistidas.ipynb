```python
#TOTAL PAGADO A CIERTO MES
query=f'''
WITH BASE AS
(SELECT SUBSCRIBER_ID
FROM DWH_BODEGA_WOM.FCT_SUBSCRIBERS_ENDING A 
WHERE A.PERIODO_PROCESO_CODIGO=202304
AND   A.SERVICIO='Prepaid'
AND   A.MOVIMIENTO_TIPO_NOMBRE='CHURN MIGRACION'),
BASE2 AS(
SELECT A.SUBS_ID,
       -1*SUM(A.PAGO) AS PAGOS,
       COUNT(*) AS FACTURAS_PAGADAS
FROM CR_DL_PAGOS A
WHERE A.SUBS_ID IN (
SELECT SUBSCRIBER_ID FROM DWH_BODEGA_WOM.FCT_SUBSCRIBERS_ENDING A
WHERE A.PERIODO_PROCESO_CODIGO=202304
AND   A.SERVICIO='Prepaid'
AND   A.MOVIMIENTO_TIPO_NOMBRE='CHURN MIGRACION')
AND SALDO=0
AND TO_NUMBER(TO_CHAR(FECHA_PAGO,'YYYYMMDD'))<20230505
GROUP BY SUBS_ID )


SELECT SUM(PAGOS) AS PAGO_TOTAL,
       SUM(FACTURAS_PAGADAS) AS TOTAL
FROM(
SELECT A.SUBSCRIBER_ID,
       COALESCE(B.PAGOS,0) AS PAGOS,
       COALESCE(B.FACTURAS_PAGADAS,0) AS FACTURAS_PAGADAS,
       ROUND(COALESCE(B.PAGOS,0)/COALESCE(B.FACTURAS_PAGADAS,1),0) AS PROMEDIO_FACTURA
       
FROM BASE A 
LEFT JOIN BASE2 B ON A.SUBSCRIBER_ID=B.SUBS_ID
)
'''
```

```python
#procedimiento mccm 
query=f'''
create or replace PROCEDURE PROC_PREPAGO_MCCM
AS
BEGIN 
    --------------------------------------------
    -- Tabla del MCCM---------------------------
    --------------------------------------------
    DELETE FROM CR_DL_PREPAGO_MCCM;
    COMMIT;
    INSERT /*+APPEND*/
    INTO CR_DL_PREPAGO_MCCM nologging
        SELECT /*+PARALLEL(8)*/ 
            TO_NUMBER(TO_CHAR(SYSTIMESTAMP, 'YYYYMMDDHH')) AS ULT_ACTUALIZACION,
            A.SUBSCRIBER_ID,
            A.MSISDN,
            TO_DATE(TO_CHAR(SYSTIMESTAMP, 'YYYY-MM-DD'), 'YYYY-MM-DD') -  TO_DATE(TO_CHAR(A.FECHA_ALTA, 'YYYY-MM-DD'), 'YYYY-MM-DD') AS DIAS_ANTIGUEDAD,
            A.ESTADO,
            COALESCE(A.OPERADOR_PORTA_ORIGEN, 'LINEA_NUEVA') AS OPERADOR_ORIGEN,
            B.CONSULTOR_REGIONAL_NOMBRE AS DEALER_REGIONAL,
            B.CONSULTOR_NOMBRE_NIVEL_8 AS DEALER_NOMBRE,
            TO_DATE(TO_CHAR(SYSTIMESTAMP, 'YYYYMMDD'), 'YYYYMMDD') - CASE
            WHEN J.ULT_TRAFICO = 999 THEN (TO_DATE(20200101, 'YYYYMMDD'))        
            WHEN J.ULT_TRAFICO <> 0 THEN (TO_DATE(J.ULT_TRAFICO, 'YYYYMMDD'))
            ELSE TO_DATE(TO_CHAR(SYSTIMESTAMP-91, 'YYYYMMDD'), 'YYYYMMDD')
            END DIAS_INACT,
            CASE
                WHEN A.MOVIMIENTO_TIPO_NOMBRE IN ('ACTIVACION NO GROSS', 'PERMANECE NO GROSS') THEN '0.Activacion'
                ELSE C.SEGMENTACION_RFM 
            END SEGMENTACION,
            M.SEG_RENTABILIZACION AS SUB_SEGMENTACION_1,
            A.MOVIMIENTO_TIPO_NOMBRE  AS SUB_SEGMENTACION_2,
            --'<--->' AS SUB_SEGMENTACION_3,
            O.TIPO_CONSUMO AS SUB_SEGMENTACION_3,
            '<--->' AS SUB_SEGMENTACION_4,
            '<--->' AS SUB_SEGMENTACION_5,
            D.PORTOUT AS CHURN_PORTOUT,
            D.INACTIVIDAD AS CHURN_INACTIVIDAD,
            D.TIPO_USUARIO,
            D.CONSUMO AS TIPO_TRAFICO,
            D.DEPARTAMENTO AS UBIC_DEPARTAMENTO, 
            D.MUNICIPIO AS UBIC_MUNICIPIO, 
            D.LOCALIDAD AS UBIC_LOCALIDAD, 
            D.BARRIO AS UBIC_BARRIO,
            G.TIMESTAMP AS FRANJA_HORARIA_CONSUMO,
            SUBSTR(F.DEVICE_MARCA, 1, 30) AS DEVICE_MARCA,
            SUBSTR(F.DEVICE_MODELO, 1, 30) AS DEVICE_MODELO,
            SUBSTR(F.DEVICE_CANT_SIMS, 1, 30) AS DEVICE_CANT_SIMS,
            F.COMPATIBILIDAD_3G AS DEVICE_COMP_3G,
            F.COMPATIBILIDAD_4G AS DEVICE_COMP_4G,
            F.COMPATIBILIDAD_WOM AS DEVICE_COMP_WOM,
            E.VALOR_RECARGAS_M1 AS RECARGAS_VALOR_temp1,
            E.TOT_RECARGAS_M1 AS RECARGAS_CANT_temp1,
            E.VALOR_RECARGAS_M3 AS RECARGAS_VALOR_temp2,
            E.TOT_RECARGAS_M3 AS RECARGAS_CANT_temp2,
            E.ULT_RECARGA AS ULTIMA_RECARGA,
            E.VALOR_PAQUETES_M1 AS PAQUETES_VALOR_temp1,
            E.TOT_PAQUETES_M1 AS PAQUETES_CANT_temp1,
            E.VALOR_PAQUETES_M3 AS PAQUETES_VALOR_temp2,
            E.TOT_PAQUETES_M3 AS PAQUETES_CANT_temp2,
            SUBSTR(E.ULT_PAQUETE, 1, 30) AS ULTIMO_PAQUETE,
            SUBSTR(H.FAV_PAQUETE_1M, 1, 30) AS FAV_PAQUETE_temp1 ,
            H.CANT_FAV_PAQUETE_1M AS CANT_FAV_PAQUETE_temp1,
            SUBSTR(H.FAV_PAQUETE_3M, 1, 30) AS FAV_PAQUETE_temp2 ,
            H.CANT_FAV_PAQUETE_3M AS CANT_FAV_PAQUETE_temp2,
            I.CANAL_CONTACTO_VALOR,
            I.VALOR_CANAL,
            I.CANAL_CONTACTO_CANT,
            I.CANT_CANAL,
            K.FRANJA_HORARIA AS RECARGAS_FRANJA_HORARIA,
            'Prepago' AS LABEL_1,
            SUBSTR(L.PAQUETE_NOMBRE, 1, 30) AS LABEL_2,
            N.REGISTRO AS LABEL_3,
            --COALESCE(TO_CHAR(S.BOOLEAN_SIEMBRA), '0') AS LABEL_4,
            T.GRUPO_TRATAMIENTO AS LABEL_4,
            '<--->' AS LABEL_5,
            TO_NUMBER(TO_CHAR(L.FECHA_RECARGA, 'YYYYMMDD'))AS NUMERIC_VALUE_1,
            TO_NUMBER(TO_CHAR(L.FECHA_EXPIRACION, 'YYYYMMDD'))AS NUMERIC_VALUE_2,
            P.PREDICCION_GQ AS NUMERIC_VALUE_3,
            T.VALOR_CARGA AS NUMERIC_VALUE_4,
            -2 AS NUMERIC_VALUE_5
        FROM DWH_BODEGA_WOM.FCT_SUBSCRIBERS_ENDING A
        LEFT JOIN DWH_BODEGA_WOM.DIM_CONSULTORES B ON A.CONSULTOR_DK = B.CONSULTOR_DK
        LEFT JOIN SEGMENTACION_RFM C ON A.SUBSCRIBER_ID = C.SUBSCRIBER_ID 
        LEFT JOIN CR_DL_INACTIVIDADES_PRE J ON A.SUBSCRIBER_ID = J.SUBSCRIBER_ID
        --LEFT JOIN (SELECT * FROM MODELO_CHURN_PREPAGO WHERE SEMANA = (TO_NUMBER(TO_CHAR(SYSTIMESTAMP, 'IW')))) D ON A.SUBSCRIBER_ID = D.SUBSCRIBER_ID
        LEFT JOIN (
            SELECT 
                SUBSCRIBER_ID,
                PORTOUT,
                INACTIVIDAD,
                TIPO_USUARIO,
                CONSUMO,
                DEPARTAMENTO, 
                MUNICIPIO, 
                LOCALIDAD, 
                BARRIO
            FROM (
                SELECT 
                    SUBSCRIBER_ID,
                    PORTOUT,
                    INACTIVIDAD,
                    TIPO_USUARIO,
                    CONSUMO,
                    DEPARTAMENTO, 
                    MUNICIPIO, 
                    LOCALIDAD, 
                    BARRIO,
                    ROW_NUMBER() OVER (PARTITION BY SUBSCRIBER_ID ORDER BY SEMANA DESC) AS RN
                FROM MODELO_CHURN_PREPAGO 
            WHERE SEMANA = 44)
            WHERE RN = 1
        ) D ON A.SUBSCRIBER_ID = D.SUBSCRIBER_ID
        LEFT JOIN AGG_WL_FRANJA_USUARIO G ON A.MSISDN = G.NUM_CELULAR
        LEFT JOIN AGG_DL_ENDING_DEVICES F ON A.SUBSCRIBER_ID = F.SUBSCRIBER_ID
        LEFT JOIN AGG_DL_RECARGAS_PAQUETES E ON A.SUBSCRIBER_ID = E.SUBSCRIBER_ID
        LEFT JOIN AGG_DL_FRECUENCIA_PAQUETES H ON A.SUBSCRIBER_ID = H.SUBSCRIBER_ID
        LEFT JOIN AGG_DL_GESTORES_RECARGAS I ON A.SUBSCRIBER_ID = I.SUBSCRIBER_ID
        LEFT JOIN AGG_DL_FRANJA_RECARGAS K ON A.SUBSCRIBER_ID = K.SUBSCRIBER_ID
        LEFT JOIN (
            SELECT /*+PARALLEL(2)*/
                SUBSCRIBER_ID,
                PAQUETE_NOMBRE,
                FECHA_RECARGA,
                FECHA_EXPIRACION
                FROM (
                    SELECT /*+PARALLEL(2)*/
                        SUBSCRIBER_ID,
                        PAQUETE_NOMBRE,
                        FECHA_RECARGA,
                        FECHA_EXPIRACION,
                        ROW_NUMBER() OVER (PARTITION BY SUBSCRIBER_ID ORDER BY FECHA_EXPIRACION DESC) AS RN
                    FROM DWH_BODEGA_WOM.FCT_PAQUETES
                    WHERE FECHA_EXPIRACION >= SYSTIMESTAMP
                        AND SERVICIO = 'Prepaid')
            WHERE RN = 1
            )L ON A.SUBSCRIBER_ID = L.SUBSCRIBER_ID
        LEFT JOIN (
            SELECT /*+PARALLEL(2)*/
                SUBSCRIBER_ID,
                SEG_RENTABILIZACION
            FROM MVALIENTE.TBL_ANL_BASE_PREPAGO
            --WHERE PERIODO = TO_NUMBER(TO_CHAR(SYSTIMESTAMP, 'YYYYMM'))        
            WHERE PERIODO = 202304 
        ) M ON A.SUBSCRIBER_ID = M.SUBSCRIBER_ID
        LEFT JOIN (
            SELECT 
                SUBSCRIBER_ID,
                REGISTRO
            FROM AGG_DL_REGISTRO_PREPAGO
        )N ON A.SUBSCRIBER_ID = N.SUBSCRIBER_ID
        LEFT JOIN AGG_DL_TRAFICO_30_PRE O ON A.SUBSCRIBER_ID = O.SUBSCRIBER_ID
        LEFT JOIN (
            SELECT *
            FROM (
            SELECT 
                SUBSCRIBER_ID,
                PREDICCION AS PREDICCION_GQ,
                ROW_NUMBER() OVER (PARTITION BY SUBSCRIBER_ID ORDER BY PERIODO_PROCESO_CODIGO) AS RN
            FROM AGG_DL_PJ_QUALITY
            )
            WHERE RN = 1
        )P ON A.SUBSCRIBER_ID = P.SUBSCRIBER_ID
        LEFT JOIN (
            SELECT *
            FROM (
                SELECT 
                    SUBSCRIBER_ID,
                    PREDICCION AS PREDICCION_CP,
                    ROW_NUMBER() OVER (PARTITION BY SUBSCRIBER_ID ORDER BY PERIODO_PROCESO_CODIGO) AS RN
                FROM AGG_DL_PJ_CHURN_PREDICTOR
            )
            WHERE RN = 1
        )Q ON A.SUBSCRIBER_ID = Q.SUBSCRIBER_ID
        LEFT JOIN (
            SELECT 
                SUBSCRIBER_ID,
                PREDICCION AS PREDICCION_TU,
                VALOR_CARGA
            FROM AGG_DL_PJ_TOPUPS
            WHERE FECHA_PRED = TO_NUMBER(TO_CHAR(SYSDATE, 'YYYYMMDD'))
        )R ON A.SUBSCRIBER_ID = R.SUBSCRIBER_ID
        --LEFT JOIN (
            --SELECT  
                --CASE 
                    --WHEN LENGTH(MSISDN) = 10 THEN MSISDN
                    --WHEN LENGTH(MSISDN) = 12 THEN SUBSTR(MSISDN, 3, 10)
                --END NUMERO,
                --CASE
                    --WHEN FECHA_ULTIMO_REGISTRO IS NULL THEN 0
                    --ELSE 1
                --END BOOLEAN_SIEMBRA
            --FROM MVALIENTE.DROPME_ENCENDIDOS
        --)S ON TO_CHAR(A.MSISDN) = TO_CHAR(S.NUMERO)
        LEFT JOIN (
            SELECT *
            FROM(
                SELECT 
                    SUBSCRIBER_ID,
                    VALOR_CARGA,
                    GRUPO_TRATAMIENTO,
                    ROW_NUMBER() OVER (PARTITION BY SUBSCRIBER_ID ORDER BY VALOR_CARGA DESC) AS RN
                FROM AGG_DL_PJ_TOPUPS_INS_GEST
                WHERE PERIODO_PROCESO_CODIGO = TO_NUMBER(TO_CHAR(SYSDATE, 'YYYYMMDD'))
                    AND GRUPO_TRATAMIENTO LIKE '%Tratamiento%'
            )
            WHERE RN = 1
        )T ON A.SUBSCRIBER_ID = T.SUBSCRIBER_ID
        WHERE A.PERIODO_PROCESO_CODIGO = TO_NUMBER(TO_CHAR(SYSTIMESTAMP, 'YYYYMM'))
        --WHERE A.PERIODO_PROCESO_CODIGO = 202209

            AND A.SERVICIO = 'Prepaid';
        COMMIT;
END;
'''
```
